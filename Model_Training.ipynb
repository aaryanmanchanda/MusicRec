Text Cleaning/ Text Preprocessing

```python
df['text'] = df['text'].str.lower().replace(r'^\w\s', ' ').replace(r'\n', ' ', regex = True)
```

```python
df.tail(5)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
nltk.download('punkt')
stemmer = PorterStemmer()

def tokenization(txt):
    tokens = nltk.word_tokenize(txt)
    stemming = [stemmer.stem(w) for w in tokens]
    return " ".join(stemming)
```

```python
df['text'] = df['text'].apply(lambda x: tokenization(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidvector = TfidfVectorizer(analyzer='word',stop_words='english')
matrix = tfidvector.fit_transform(df['text'])
similarity = cosine_similarity(matrix)
```

```python
# def recommendation(song_df):
#     idx = df[df['song'] == song_df].index[0]
#     distances = sorted(list(enumerate(similarity[idx])),reverse=True,key=lambda x:x[1])

#     songs = []
#     for m_id in distances[1:21]:
#         songs.append(df.iloc[m_id[0]].song)

#     return songs
def recommendation(song_df, df):
  """Recommends similar songs based on a song in the DataFrame.

  Args:
      song_df: The song title to find recommendations for.
      df: The DataFrame containing song data and similarity information.

  Returns:
      A list of recommended songs (up to 20) or an empty list if no song is found.
  """
  if len(df) == 0:
    return []  # Return empty list if DataFrame is empty

  matching_songs = df[df['song'] == song_df]
  if matching_songs.empty:
    return []  # Return empty list if no matching song is found

  idx = matching_songs.index[0]  # Get index of the first matching song
  distances = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda x: x[1])

  songs = []
  for m_id in distances[1:21]:
    songs.append(df.iloc[m_id[0]].song)

  return songs
```

```python
recommendation('We\'re All Alone',df)
```
